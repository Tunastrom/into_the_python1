```python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
writing_list = []
company = input('input company: ')
print()
```

```python
data = yf.download(company, start = '2021-06-01')
print(data)
```

```python
type(data)
```

```python
plt.plot(data['Open'],color='forestgreen')
plt.plot(data.index,data['Close'],color='darkorange')
plt.ylabel('Price')
plt.xlabel('Date')
plt.grid(True)
plt.title(f'{company} Open/Close')
plt.show()
```

```python
plt.plot(data.index,data['High'],color='salmon')   
plt.plot(data.index,data['Low'], color='royalblue')
plt.ylabel('Price')
plt.xlabel('Date')
plt.grid(True)
plt.title(f'{company} High/Low')
plt.show()
```
