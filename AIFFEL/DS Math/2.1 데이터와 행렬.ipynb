#### 연습문제 2.1.1

````python
from sklearn.datasets import load_digits # 패키지 임포트

digits = load_digits() # 데이터 로드
samples = [0, 10, 20, 30, 1, 11, 21,31]
d = []
for i in range(8):
    d.append(digits.images[samples[i]])

plt.figure(figsize=(8,2))
for i in range(8):
    plt.subplot(1, 8, i + 1)
    plt.imshow(d[i], interpolation='nearest', cmap=plt.cm.bone_r)
    plt.grid(False); plt.xticks([]); plt.yticks([])
    plt.title("image {}".format(i + 1))
plt.suptitle("number 0 and 1 images")
plt.tight_layout()
plt.show()
````

#### 64크기의 1차원 벡터로 펼치기

````python
v = []
for i in range(8):
    v.append(d[i].reshape(64,1))   # 벡터화

plt.figure(figsize=(8,2))
for i in range(8):
    plt.subplot(1, 8, i + 1)
    plt.imshow(v[i], aspect=0.4, interpolation='nearest', cmap=plt.cm.bone_r)
    plt.grid(False); plt.xticks([]); plt.yticks([])
    plt.title("image {}".format(i + 1))
plt.suptitle("vetctorized images", y=1.05)
plt.tight_layout()
plt.show()
````

### 컬러이미지 예제

````python
from scipy import misc # 패키지 임포트

img_rgb = misc.face() # 컬러이미지 로드
img_rgb.shape # 데이터의 모양
````

````python
plt.subplot(221)
plt.imshow(img_rgb, cmap=plt.cm.gray)  # 컬러 이미지 출력
plt.axis("off")
plt.title("RGB color image")

plt.subplot(222)
plt.imshow(img_rgb[:, :, 0], cmap=plt.cm.gray) # red 채널 출력
plt.axis("off")
plt.title("Red channel")

plt.subplot(223)
plt.imshow(img_rgb[:, :, 1], cmap=plt.cm.gray) # Green 채널 출력
plt.axis("off")
plt.title("Green channel")

plt.subplot(224)
plt.imshow(img_rgb[:, :, 2], cmap=plt.cm.gray) # Blue 채널 출력
plt.axis("off")
plt.title("Blue channel")

plt.show()
````

#### 위 RGB 컬러 이미지는 2차원의 행렬일 것 같지만,사실 red, green, blue 세 개의 채널(행렬)이 겹쳐진 텐서이다.
**채널(channel)**: 컬러이미지에서 각각의 색을 나타내는 행렬
### 예제 이미지의 크기가 768 x 1024 이므로 768 x 1024 x 3 크기의 3차원 텐서이다.
